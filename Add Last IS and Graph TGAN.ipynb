```python
import torch
import numpy as np
from models.tgan import VideoGenerator
from evaluation_metrics import calculate_inception_score

path = 'ucf101/tgan'
scores = np.load('tgan_inception.npy')
print(scores)
```

```python
# gen model
gen = VideoGenerator()
state_dicts = torch.load(f'checkpoints/{path}/state_normal99999.ckpt')

gen.load_state_dict(state_dicts['model_state_dict'][0])
gen.cpu()
results = calculate_inception_score(gen, test=False, zdim=100)
print(results)
```

```python
scores = list(scores)
scores.append(results)
```

```python
np.save('tgan_inception.npy', scores)
```
